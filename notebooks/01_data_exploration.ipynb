```python
import os
import matplotlib.pyplot as plt
from collections import Counter
from PIL import Image
```

```python
def count_images(base_path):
    counts = {}
    for split in ["train", "val", "test"]:
        counts[split] = {
            cls: len(os.listdir(os.path.join(base_path, split, cls)))
            for cls in ["cats", "dogs"]
        }
    return counts

counts = count_images("../data")
counts
```

```python
for split, data in counts.items():
    plt.figure()
    plt.bar(data.keys(), data.values())
    plt.title(f"{split.capitalize()} Class Distribution")
    plt.ylabel("Number of Images")
    plt.show()
```

```python
def show_samples(split="train", cls="cats", n=5):
    folder = f"../data/{split}/{cls}"
    images = os.listdir(folder)[:n]

    plt.figure(figsize=(15,3))
    for i, img_name in enumerate(images):
        img = Image.open(os.path.join(folder, img_name))
        plt.subplot(1, n, i+1)
        plt.imshow(img)
        plt.axis("off")
    plt.suptitle(f"Sample {cls} images ({split})")
    plt.show()

show_samples("train", "cats")
show_samples("train", "dogs")
```

Dataset Overview
The Kaggle Dogs vs Cats dataset was selected due to its balanced binary classes and high-quality annotations. The dataset was split into training (70%), validation (15%), and test (15%) sets to enable unbiased model evaluation. Exploratory analysis confirms balanced class distributions across splits, making it suitable for transfer learning experiments.
